```python
titanic = pd.read_csv('data/raw/titanic.csv')
```

```python
model = RandomForestClassifier(random_state=42)
```

```python
X = titanic.iloc[:,1:]
sex = pd.get_dummies(X['Sex'])
X.drop(['Name', 'Sex'], axis=1, inplace=True)
X = pd.concat([X, sex], axis=1)
y = titanic.iloc[:,0]
```

```python
model.fit(X, y)
```

```python
joblib.dump(model, '')
```
